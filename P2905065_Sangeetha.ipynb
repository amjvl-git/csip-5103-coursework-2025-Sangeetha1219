```python
import torch
import torch.nn as nn
import torch.optim as optim
import torch.nn.functional as F
from torch.utils.data import DataLoader, random_split
import torchvision
import torchvision.transforms as transforms
from torch.optim.lr_scheduler import StepLR
import matplotlib.pyplot as plt
import numpy as np
from sklearn.metrics import confusion_matrix, classification_report
from sklearn.model_selection import ParameterGrid
import itertools
import copy

# Set device
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print(f"Using device: {device}")

# Step 1: Load the Dataset with Enhanced Preprocessing
# Enhanced data augmentation for training (RandomCrop, GaussianBlur for grayscale)
train_transform = transforms.Compose([
    transforms.RandomHorizontalFlip(p=0.5),
    transforms.RandomRotation(10),
    transforms.RandomCrop(28, padding=2),  # RandomCrop with padding
    transforms.GaussianBlur(kernel_size=3, sigma=(0.1, 2.0)),  # Blur for robustness (grayscale-friendly)
    transforms.ToTensor(),
    transforms.Normalize(mean=[0.2860], std=[0.3530])  # Fashion-MNIST stats
])

test_transform = transforms.Compose([
    transforms.ToTensor(),
    transforms.Normalize(mean=[0.2860], std=[0.3530])
])

# Load datasets
train_dataset = torchvision.datasets.FashionMNIST(root='./data', train=True, download=True, transform=train_transform)
test_dataset = torchvision.datasets.FashionMNIST(root='./data', train=False, download=True, transform=test_transform)

# Split train into train (80%) and val (20%)
train_size = int(0.8 * len(train_dataset))
val_size = len(train_dataset) - train_size
train_ds, val_ds = random_split(train_dataset, [train_size, val_size])

# DataLoaders
batch_size = 64
train_loader = DataLoader(train_ds, batch_size=batch_size, shuffle=True)
val_loader = DataLoader(val_ds, batch_size=batch_size, shuffle=False)
test_loader = DataLoader(test_dataset, batch_size=batch_size, shuffle=False)

print(f"Train: {len(train_ds)}, Val: {len(val_ds)}, Test: {len(test_dataset)}")

# Class names
classes = ['T-shirt/top', 'Trouser', 'Pullover', 'Dress', 'Coat', 'Sandal', 'Shirt', 'Sneaker', 'Bag', 'Ankle boot']

# Step 2: Enhanced Model Design (Deeper CNN with more conv layers)
class EnhancedANN(nn.Module):
    def __init__(self, dropout_rate=0.5):
        super(EnhancedANN, self).__init__()
        self.flatten = nn.Flatten()
        self.fc1 = nn.Linear(28*28, 512)
        self.dropout1 = nn.Dropout(dropout_rate)
        self.fc2 = nn.Linear(512, 256)
        self.dropout2 = nn.Dropout(dropout_rate)
        self.fc3 = nn.Linear(256, 10)
        self.relu = nn.ReLU()

    def forward(self, x):
        x = self.flatten(x)
        x = self.relu(self.fc1(x))
        x = self.dropout1(x)
        x = self.relu(self.fc2(x))
        x = self.dropout2(x)
        x = self.fc3(x)
        return x

class DeeperCNN(nn.Module):
    def __init__(self, dropout_rate=0.25):
        super(DeeperCNN, self).__init__()
        self.conv1 = nn.Conv2d(1, 32, kernel_size=3, padding=1)
        self.conv2 = nn.Conv2d(32, 64, kernel_size=3, padding=1)
        self.conv3 = nn.Conv2d(64, 128, kernel_size=3, padding=1)  # Additional deeper layer
        self.pool = nn.MaxPool2d(2, 2)
        self.dropout = nn.Dropout(dropout_rate)
        self.fc1 = nn.Linear(128 * 3 * 3, 256)  # After three pools: 28/8=3.5 -> 3
        self.fc2 = nn.Linear(256, 10)
        self.relu = nn.ReLU()

    def forward(self, x):
        x = self.pool(self.relu(self.conv1(x)))  # 28->14
        x = self.pool(self.relu(self.conv2(x)))  # 14->7
        x = self.pool(self.relu(self.conv3(x)))  # 7->3
        x = x.view(-1, 128 * 3 * 3)
        x = self.relu(self.fc1(x))
        x = self.dropout(x)
        x = self.fc2(x)
        return x

# Ensemble Model (averages softmax predictions from ANN and CNN)
class EnsembleModel(nn.Module):
    def __init__(self, ann_model, cnn_model):
        super(EnsembleModel, self).__init__()
        self.ann = ann_model
        self.cnn = cnn_model

    def forward(self, x):
        ann_out = F.softmax(self.ann(x), dim=1)
        cnn_out = F.softmax(self.cnn(x), dim=1)
        ensemble_out = (ann_out + cnn_out) / 2
        return torch.log(ensemble_out + 1e-8)  # Log for CrossEntropyLoss

# Hyperparameter Tuning (Grid search on LR and dropout)
param_grid = {
    'lr': [0.001, 0.0001],
    'dropout': [0.3, 0.5]
}

def grid_search_tune(model_class, train_loader, val_loader, param_grid, epochs=5, name="Model"):
    best_val_acc = 0
    best_params = None
    best_model = None

    for params in ParameterGrid(param_grid):
        print(f"Tuning {name} with params: {params}")
        model = model_class(dropout_rate=params['dropout']).to(device)
        optimizer = optim.Adam(model.parameters(), lr=params['lr'])
        criterion = nn.CrossEntropyLoss()
        scheduler = StepLR(optimizer, step_size=3, gamma=0.1)

        # Quick training for tuning (fewer epochs)
        model.train()
        for epoch in range(epochs):
            for inputs, labels in train_loader:
                inputs, labels = inputs.to(device), labels.to(device)
                optimizer.zero_grad()
                outputs = model(inputs)
                loss = criterion(outputs, labels)
                loss.backward()
                optimizer.step()
            scheduler.step()

        # Quick val eval
        model.eval()
        correct, total = 0, 0
        with torch.no_grad():
            for inputs, labels in val_loader:
                inputs, labels = inputs.to(device), labels.to(device)
                outputs = model(inputs)
                _, predicted = torch.max(outputs, 1)
                total += labels.size(0)
                correct += (predicted == labels).sum().item()
        val_acc = 100 * correct / total

        print(f"Val Acc: {val_acc:.2f}%")
        if val_acc > best_val_acc:
            best_val_acc = val_acc
            best_params = params
            best_model = copy.deepcopy(model)

    print(f"Best {name} params: {best_params}, Val Acc: {best_val_acc:.2f}%")
    return best_model, best_params

# Tune and get best models
print("Tuning ANN...")
best_ann, best_ann_params = grid_search_tune(EnhancedANN, train_loader, val_loader, param_grid, epochs=5, name="ANN")

print("Tuning CNN...")
best_cnn, best_cnn_params = grid_search_tune(DeeperCNN, train_loader, val_loader, param_grid, epochs=5, name="CNN")

# Ensemble
ensemble_model = EnsembleModel(best_ann, best_cnn).to(device)

# Hyperparameters for full training (using best from tuning)
lr = best_ann_params['lr']  # Use ANN's best LR for consistency
epochs = 10

optimizer_ensemble = optim.Adam(ensemble_model.parameters(), lr=lr)
scheduler_ensemble = StepLR(optimizer_ensemble, step_size=7, gamma=0.1)
criterion = nn.CrossEntropyLoss()

# Training function (updated for ensemble)
def train_model(model, train_loader, val_loader, optimizer, scheduler, criterion, epochs, name):
    train_losses, val_losses = [], []
    train_accs, val_accs = [], []

    for epoch in range(epochs):
        # Train
        model.train()
        running_loss, correct, total = 0.0, 0, 0
        for inputs, labels in train_loader:
            inputs, labels = inputs.to(device), labels.to(device)
            optimizer.zero_grad()
            outputs = model(inputs)
            loss = criterion(outputs, labels)
            loss.backward()
            optimizer.step()

            running_loss += loss.item()
            _, predicted = torch.max(outputs, 1)
            total += labels.size(0)
            correct += (predicted == labels).sum().item()

        train_loss = running_loss / len(train_loader)
        train_acc = 100 * correct / total
        train_losses.append(train_loss)
        train_accs.append(train_acc)

        # Validate
        model.eval()
        val_loss, correct, total = 0.0, 0, 0
        with torch.no_grad():
            for inputs, labels in val_loader:
                inputs, labels = inputs.to(device), labels.to(device)
                outputs = model(inputs)
                loss = criterion(outputs, labels)
                val_loss += loss.item()
                _, predicted = torch.max(outputs, 1)
                total += labels.size(0)
                correct += (predicted == labels).sum().item()

        val_loss_avg = val_loss / len(val_loader)
        val_acc = 100 * correct / total
        val_losses.append(val_loss_avg)
        val_accs.append(val_acc)

        scheduler.step()
        print(f"{name} Epoch {epoch+1}/{epochs}: Train Loss={train_loss:.4f}, Acc={train_acc:.2f}% | Val Loss={val_loss_avg:.4f}, Acc={val_acc:.2f}%")

    # Plot curves
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 4))
    ax1.plot(train_losses, label='Train Loss')
    ax1.plot(val_losses, label='Val Loss')
    ax1.set_title(f'{name} Loss Curves')
    ax1.legend()

    ax2.plot(train_accs, label='Train Acc')
    ax2.plot(val_accs, label='Val Acc')
    ax2.set_title(f'{name} Accuracy Curves')
    ax2.legend()
    plt.show()

    return train_losses, val_losses, train_accs, val_accs

# Train individual models fully (using best params)
print("Full Training ANN...")
ann_optimizer = optim.Adam(best_ann.parameters(), lr=lr)
ann_scheduler = StepLR(ann_optimizer, step_size=7, gamma=0.1)
ann_train_losses, ann_val_losses, ann_train_accs, ann_val_accs = train_model(best_ann, train_loader, val_loader, ann_optimizer, ann_scheduler, criterion, epochs, "ANN")

print("\nFull Training CNN...")
cnn_optimizer = optim.Adam(best_cnn.parameters(), lr=lr)
cnn_scheduler = StepLR(cnn_optimizer, step_size=7, gamma=0.1)
cnn_train_losses, cnn_val_losses, cnn_train_accs, cnn_val_accs = train_model(best_cnn, train_loader, val_loader, cnn_optimizer, cnn_scheduler, criterion, epochs, "CNN")

print("\nFull Training Ensemble...")
ensemble_train_losses, ensemble_val_losses, ensemble_train_accs, ensemble_val_accs = train_model(ensemble_model, train_loader, val_loader, optimizer_ensemble, scheduler_ensemble, criterion, epochs, "Ensemble")

# Evaluation function
def evaluate_model(model, test_loader, name):
    model.eval()
    all_preds, all_labels = [], []
    correct, total = 0, 0
    with torch.no_grad():
        for inputs, labels in test_loader:
            inputs, labels = inputs.to(device), labels.to(device)
            outputs = model(inputs)
            _, predicted = torch.max(outputs, 1)
            total += labels.size(0)
            correct += (predicted == labels).sum().item()
            all_preds.extend(predicted.cpu().numpy())
            all_labels.extend(labels.cpu().numpy())

    test_acc = 100 * correct / total
    print(f"{name} Test Accuracy: {test_acc:.2f}%")

    # Confusion Matrix
    cm = confusion_matrix(all_labels, all_preds)
    plt.figure(figsize=(10, 8))
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=classes, yticklabels=classes)
    plt.title(f'{name} Confusion Matrix')
    plt.xlabel('Predicted')
    plt.ylabel('True')
    plt.show()

    # Class-wise report
    print(classification_report(all_labels, all_preds, target_names=classes))

    return test_acc, cm

# Evaluate all
ann_acc, ann_cm = evaluate_model(best_ann, test_loader, "ANN")
cnn_acc, cnn_cm = evaluate_model(best_cnn, test_loader, "CNN")
ensemble_acc, ensemble_cm = evaluate_model(ensemble_model, test_loader, "Ensemble")

# Analysis and Comparison
print("\n--- Analysis ---")
print(f"ANN Test Accuracy: {ann_acc:.2f}% (Params: {best_ann_params})")
print(f"CNN Test Accuracy: {cnn_acc:.2f}% (Params: {best_cnn_params})")
print(f"Ensemble Test Accuracy: {ensemble_acc:.2f}% (Averaging boosts by {ensemble_acc - max(ann_acc, cnn_acc):.2f}%)")

# Highlight confusions
shirt_coat_conf_ann = ann_cm[6,4] + ann_cm[4,6]
print(f"ANN Shirt (6) vs. Coat (4) confusions: {shirt_coat_conf_ann} (visually similar; deeper CNN reduces this).")

shirt_coat_conf_cnn = cnn_cm[6,4] + cnn_cm[4,6]
print(f"CNN Shirt vs. Coat confusions: {shirt_coat_conf_cnn}.")

# Save models
torch.save(best_ann.state_dict(), 'best_ann_fashion_mnist.pth')
torch.save(best_cnn.state_dict(), 'best_cnn_fashion_mnist.pth')
torch.save(ensemble_model.state_dict(), 'ensemble_fashion_mnist.pth')
print("Best models saved.")
```
